```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Purchase Dataset Experiments

```python
# 1. Affect of Sharding on the Purchase Dataset
Data_1 = {'percantage_data_unlearn_req': [20,40,60,80],
        'classification_accuracy': [98.200,98.200,98.200,98.200]
       }
  
df_1 = pd.DataFrame(Data_1,columns=['percantage_data_unlearn_req','classification_accuracy'])

Data_3 = {'percantage_data_unlearn_req': [20,40,60,80],
        'classification_accuracy': [98.3500,98.400,98.400,98.400]
       }
  
df_3 = pd.DataFrame(Data_3,columns=['percantage_data_unlearn_req','classification_accuracy'])

Data_5 = {'percantage_data_unlearn_req': [20,40,60,80],
        'classification_accuracy': [98.050,98.100,98.100,98.100]
       }
  
df_5 = pd.DataFrame(Data_5,columns=['percantage_data_unlearn_req','classification_accuracy'])

Data_7 = {'percantage_data_unlearn_req': [20,40,60,80],
        'classification_accuracy': [97.900,98.000,98.050,98.050]
       }
  
df_7 = pd.DataFrame(Data_7,columns=['percantage_data_unlearn_req','classification_accuracy'])

Data_9 = {'percantage_data_unlearn_req': [20,40,60,80],
        'classification_accuracy': [98.150,98.150,98.150,98.150]
       }
  
df_9 = pd.DataFrame(Data_9,columns=['percantage_data_unlearn_req','classification_accuracy'])

Data_base = {'percantage_data_unlearn_req': [20,40,60,80],
        'classification_accuracy': [98.200,98.300,98.450,98.250]
       }
  
df_base = pd.DataFrame(Data_base,columns=['percantage_data_unlearn_req','classification_accuracy'])
  
plt.plot(df_1['percantage_data_unlearn_req'], df_1['classification_accuracy'], color='red', marker='o', label='1 Shard, 10 slices')
plt.plot(df_3['percantage_data_unlearn_req'], df_3['classification_accuracy'], color='blue', marker='o', label='3 Shards, 10 slices')
plt.plot(df_5['percantage_data_unlearn_req'], df_5['classification_accuracy'], color='green', marker='o', label='5 Shards, 10 slices')
plt.plot(df_7['percantage_data_unlearn_req'], df_7['classification_accuracy'], color='gray', marker='o', label='7 Shards, 10 slices')
plt.plot(df_9['percantage_data_unlearn_req'], df_9['classification_accuracy'], color='purple', marker='o', label='9 Shards, 10 slices')
plt.plot(df_base['percantage_data_unlearn_req'], df_base['classification_accuracy'], color='black', marker='o', label='Naive Bayes baseline')
plt.title('Classification Accuracy Vs Unlearn Requests on Training Data', fontsize=14)
plt.xlabel("Unlearn Requests (%)", fontsize=14)
plt.ylabel("Classification Accuracy (%)", fontsize=14)
plt.grid(True)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
